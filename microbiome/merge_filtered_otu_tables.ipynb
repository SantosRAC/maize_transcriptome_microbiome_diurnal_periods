First, I (RACS) will create the feature table with day and night samples for Weighted Unifrac:

```python
import pandas as pd

# Load the TSV file into a DataFrame
otu_table_day = pd.read_csv('/home/santosrac/Projects/UGA_RACS/IntegrationMetataxExpression/SparXCC_OTUs_paper/filtered_otu_table_day_filtered_rel_abund_cv_filtered.tsv', sep='\t')
otu_table_day.set_index('OTU_ID', inplace=True)
otu_table_day.rename(columns=lambda col: f"{col}_d", inplace=True)
otu_table_night = pd.read_csv('/home/santosrac/Projects/UGA_RACS/IntegrationMetataxExpression/SparXCC_OTUs_paper/filtered_otu_table_night_filtered_rel_abund_cv_filtered.tsv', sep='\t')
otu_table_night.set_index('OTU_ID', inplace=True)
otu_table_night.rename(columns=lambda col: f"{col}_n", inplace=True)
```
